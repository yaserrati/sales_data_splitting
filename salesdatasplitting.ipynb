```python
df = pd.read_csv('sales.csv')
```

```python
df.rename(columns={'Invoice ID': 'Invoice_ID', 'Customer type': 'Customer_type', 'Product line': 'Product_line', 'Unit price': 'Unit_price', 'Tax 5%': 'Tax_5_percent', 'gross margin percentage': 'gross_margin_percentage', 'gross income': 'gross_income'}, inplace=True)
```

```python
# Shuffle the data
df_shuffled = df.sample(frac=1, random_state=42)


#input db_percentage et json_percentage
json_percentage = float(input('Enter json percentage from 0 to 1: '))
db_percentage = float(input('Enter database percentage from 0 to 1: '))

# Calculate the number of rows for each destination
total_rows = df_shuffled.shape[0]
json_rows = int(json_percentage * total_rows)
database_rows = int(db_percentage * total_rows)
csv_rows = total_rows - json_rows - database_rows

# Split the data into JSON, database, and CSV parts
df_json = df_shuffled.iloc[:json_rows]
df_database = df_shuffled.iloc[json_rows:json_rows+database_rows]
df_csv = df_shuffled.iloc[json_rows+database_rows:]
```

```python
# Save the splits to their respective destinations (JSON, database, and CSV)
df_json.to_json('json_sales.json', orient='records')
df_csv.to_csv('csv_sales.csv', index=False)
```

```python
# Créez la connexion
connection = pyodbc.connect('DRIVER={SQL SERVER} ;SERVER=LAPTOP-QJ16KG8Q\SQLEXPRESS;DATABASE=sales')

# Create a cursor to execute SQL queries
cursor = connection.cursor()
```

```python
cursor.execute('''
    CREATE TABLE [SalesTable] (
        [Invoice ID] NVARCHAR(55),
        [Branch] NVARCHAR(55),
        [City] NVARCHAR(55),
        [Customer type] NVARCHAR(55),
        [Gender] NVARCHAR(55),
        [Product line] NVARCHAR(55),
        [Unit price] FLOAT,
        [Quantity] INT,
        [Tax 5%] FLOAT,
        [Total] FLOAT,
        [Date] NVARCHAR(55),
        [Time] NVARCHAR(55),
        [Payment] NVARCHAR(55),
        [cogs] FLOAT,
        [gross margin percentage] FLOAT,
        [gross income] FLOAT,
        [Rating] FLOAT,
        PRIMARY KEY ([Invoice ID])
    );
''')
connection.commit()
```

```python
# Insert data into the dbpart table
for row in df_database.itertuples():
    # Check if the person_id already exists in the table
    cursor.execute('SELECT COUNT(*) FROM [SalesTable] WHERE [Invoice ID] = ?', (row[1],))
    count = cursor.fetchone()[0]
    
    if count == 0:
        # Insert the record if it doesn't exist
        cursor.execute('INSERT INTO SalesTable ([Invoice ID],[Branch],[City],[Customer type],[Gender],[Product line],[Unit price],[Quantity],[Tax 5%],[Total],[Date],[Time],[Payment],[cogs],[gross margin percentage],[gross income],[Rating]) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)', (row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10], row[11], row[12], row[13], row[14], row[15], row[16], row[17]))

connection.commit()
```

```python
# Close the connection
cursor.close()
connection.close()
```
